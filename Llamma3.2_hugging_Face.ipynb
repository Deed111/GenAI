```python
# loading llama-2
from transformers import AutoTokenizer
import transformers
import torch

#selected model
#model = "TinyLlama/TinyLlama-1.1B-Chat-v1.0"
model = "meta-llama/llama-2-7b-chat-hf"

tokenizer=AutoTokenizer.from_pretrained(model)
```

```python
# defining pipeline for flow of input
pipeline=transformers.pipeline(
    "text-generation",
    model=model,
   torch_dtype=torch.float16,
  device_map="auto"
)
```

```python
def get_answers(prompt):
    seq=pipeline(prompt,
                 do_sample=True,
                 return_full_text=True,
                 top_k=10,
                 num_return_sequences=1,
                 eos_token_id=tokenizer.eos_token_id,
                 )
    return seq[0]["generated_text"]
```

```python
# writing  prompt  in a defined format as required for Hf models
prompt='''
<s><INST> <SYS>
You are an authentic, honest and reliable movie recommendation system.
Your answers should be polite and respectful. If you don't know the exact answer, say so, but never give flase information or unverified information.
Always avoid spreading misinformation
</SYS>
I liked Lord of the Rings and Hobbit movies.recommend me somthing similar movies to watch
<INST>
'''
```

```python
print(get_answers(prompt))
```

```python
# trying different roles
# wrining  prompt  in a defined format as required for Hf models
prompt='''
<s><INST> <SYS>
You are an authentic, honest and reliable language professor.
Your answers should be polite and respectful. If you don't know the exact answer, say so, but never give flase information or unverified information.
Always avoid spreading misinformation
</SYS>
Tranlate following into Hindi: I am good.
<INST>
'''
```
